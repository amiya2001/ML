# 🧠 🔹 JOINS & AGGREGATION

### 1.Find the top 5 countries with the highest number of cities.

````python
pd.read_sql_query("select c.Name,count(ct.ID) as city_count from country c join city ct on c.Code=ct.CountryCode group by c.Name order by  city_count desc limit 5",conn)
````

### 2. Calculate continent-wise total population.

````python
pd.read_sql_query("select c.Continent, sum(c.Population) as total_population from country c group by c.Continent order by total_population",conn)
````

### 3.Find the country with the highest population in each continent.

````python
pd.read_sql_query("select Name,Population from country c1 where Population=(select max(Population) from country c2 where c1.Continent=c2.Continent)",conn)
````

### 4.List countries along with their capital city names.

````python
pd.read_sql_query("select c.Name ,ct.Name as capital_name from country c join city ct on c.Code=ct.CountryCode where c.Capital=ct.ID",conn)
````

# 🧠 🔹 SUBQUERIES


### 5.Find cities whose population is greater than the average city population.

````python
pd.read_sql_query("select Name from city where Population > (select avg(Population) from city )",conn)
````

### 7.Find countries whose population is greater than the average population of their continent.

````python
pd.read_sql_query("select Name,Population from country where Population > (select AVG(population) from country) ",conn)
````

# 🧠 🔹 WINDOW FUNCTIONS

### 8.Rank countries by population within each continent.


````python
pd.read_sql_query("select Name,Continent, rank() over (partition by continent order by  Population desc ) as rank from country",conn)
````

### 9.Retrieve the top 3 most populated countries in each continent.

````python
pd.read_sql_query("select * from (select Name,Continent,Population,row_number() over (partition by Continent  order by Population desc) as rnk from country ) t  where rnk<=3 ",conn)
````

### 10. Find the second most populated country in each continent.

````python
pd.read_sql_query("select * from (select Continent, Name, Population,row_number() over (partition by Continent order by population desc ) as rnk from country ) t where rnk=2",conn)
````
